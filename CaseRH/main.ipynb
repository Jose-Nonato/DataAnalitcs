```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
base_dados = pd.read_csv('Salary_Data.csv')
base_dados.head()
```

```python
# Renomeando as colunas
base_dados.rename(columns= {
    'YearsExperience': 'Xp',
    'Salary': 'Renda'
}, inplace=True)
```

```python
# Dimensão
base_dados.shape
```

```python
# Quantidade de campos nulos
base_dados.isnull().sum()
```

```python
# Gráfico de campos nulos
sns.heatmap(base_dados.isnull(), cbar=False)
```

```python
base_dados.describe()
```

```python
plt.figure(figsize=(10,6))
plt.title('Analise de Renda')
sns.kdeplot(base_dados['Renda'], color='green')
```

```python
plt.figure(figsize=(10,6))
plt.title('Analise de XP')
sns.kdeplot(base_dados['Xp'])
```

```python
# Distribuição
plt.figure(figsize=(10,6))
sns.boxplot(base_dados['Renda'])
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(base_dados['Xp'])
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=base_dados, x='Renda', y='Xp')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=base_dados, x='Renda', y='Xp')
```

```python
# Correlação
correlacao = base_dados.corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlacao, annot=True, cbar=True)
```
